Now let's put it together and loop through all the files

```python
dfs = []

for file in files:
    df = pd.read_table(file, index_col = [0,1], header=None)
    name = os.path.basename(file).split('.')[0]
    col_names = df.columns
    
    new_cols = []
    for item in col_names:
        new = name+"_"+str(item)
        new_cols.append(new)    
        
    df.columns = new_cols
    
    dfs.append(df)  
```

Example here takes all that code and puts it in a function:

```python
def load_rename_df(path_to_table):
    df = pd.read_table(path_to_table, index_col = [0,1], header=None)
    name = os.path.basename(path_to_table).split('.')[0]
    col_names = df.columns
    
    new_cols = []
    for item in col_names:
        new = name+"_"+str(item)
        new_cols.append(new)    
        
    df.columns = new_cols
    
    return df
```

```python
dfs = []

for file in files:
    df = load_rename_df(file)
    dfs.append(df)
```

Join the list of dataframes together to make one massive dataframe. In this instance, all row names are always the same (60000+ genes), so we can take the default left join. If your row names were NOT always the same, but some dataframes have additional rows, consider performing an 'outer' join by adding the argument how='outer'. See more in documentation [here](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.join.html)

```python
result = dfs[0]

for df in dfs[1:]:
    result = result.join(df)
    
print(result.shape)
result.head()
```
